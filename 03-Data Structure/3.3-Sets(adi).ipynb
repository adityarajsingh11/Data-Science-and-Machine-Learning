#### Sets
Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

```python
## create a set 
my_set = {1,2,3,4,5,6}
print(my_set)
print(type(my_set))
```

```python
## declaration of the empty set 
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_set = set([1,2,3,4,5,6])
print(my_set)
```

```python
my_empty_set = set([1,2,3,4,5,6])
print(my_empty_set)
```

```python
## Basics Sets Operation 
##Adding and removing Elements
my_set.add(7)
print(my_set)
my_set.add(7)   # duplicate element are not allowed
print(my_set)
```

```python
## remove the elements from a set 
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)
```

```python
my_set.discard(11)  # if element not available it is executed also
print(my_set)
```

```python
## Pop method 
removed_element = my_set.pop()
print(removed_element)
print(my_set)
```

```python
## Clear all elements from the sets
my_set.clear()
print(my_set)
```

```python
## Membership Operation test
my_set = {1,2,3,4}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operations

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## Union 
union_set = set1.union(set2)
print(union_set)


## Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)  ## updating common element in the set 1
print(set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## Difference 
print(set1.difference(set2))

print(set2.difference(set1))
```

```python
## Symmetric Difference 
set1.symmetric_difference(set2)  ## unique element in both set (means not intersection )
```

```python
## Sets Methods 
set1 ={1,2,3,4,5}
set2 = {3,4, 5}

## Is subset 
print(set1.issubset(set2))

## issuperset      ## means set2 ke sare element set1 se hai
print(set1.issuperset(set2))
```

```python
lst = [1,2,3,3,4,4,4,5]  ## remove the duplicate from the list 
set(lst)
```

```python
## Counting Unique words in text
text = "In this tutorial we are discussing about sets"
words = text.split()

## converts list of words to set to get unique words 

unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
