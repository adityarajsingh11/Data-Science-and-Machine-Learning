##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.



```python
# syntax 
def function_name(parameters):
    """Docstring"""
    ## function body
    return expression
```

```python
## why function?
num = 24
if num%2 ==0:
    print("the number is even")
else:
    print("the number is odd")    
```

```python
## By using the function
def even_or_odd(num):
    """This function finds the even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")  
### call the function 
even_or_odd(7)         
```

```python
## function with multiple parameters

def add(a,b):
    return a+b

result = add(4,6)
print(result)
```

```python
## Default parameters

def greet(name):
    print(f"Hello {name} welcome to gandu college")

greet("Aditya")    
```

```python
### variable length Argument 
## Positional And keyword argument 

def print_numbers(*aditya):
    for number in aditya:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,"aditya")
```

```python
## Positional arguments 
def print_numbers(*args):
    for number in args:
        print(number)

print_numbers(1,2,3,4,5,"bittu gandu")        
```

```python
### Keyword argument
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="bittu gandu",papa="prem chinnar",grandfather="Aditya")        
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument:{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")    
```

```python
print_details(1,2,3,4,name="bittu gandu",papa="prem chinnar",grandfather="Aditya") 
```

```python
## return statement 
def multiply(a,b):
    return a+b      
multiply(2,3)
```

```python
#### return multiple parameter
def multiply(a,b):
    return a*b,a

multiply(2,3)
```
